```python
import cv2
import mediapipe as mp
import math
import numpy as np
import ctypes
from ctypes import cast, POINTER
from comtypes import CLSCTX_ALL
from pycaw.pycaw import AudioUtilities, IAudioEndpointVolume
import wmi
import win32api
```

```python
# solution APIs
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_hands = mp.solutions.hands
```

```python
# Volume Control Library Usage
devices = AudioUtilities.GetSpeakers()
interface = devices.Activate(IAudioEndpointVolume._iid_, CLSCTX_ALL, None)
volume = cast(interface, POINTER(IAudioEndpointVolume))
volRange = volume.GetVolumeRange()
minVol, maxVol, volBar, volPer = volRange[0], volRange[1], 400, 0
```

```python
# Brightness Control
wCam, hCam = 1280, 720
brightness_range = [-255, 255]
minBright, maxBright, brightBar, brightPer = brightness_range[0], brightness_range[1], 150, 0
brightFactor = 1  # You can adjust this value to control the brightness sensitivity
```

```python
def set_brightness_win(brightness):
    brightness = max(min(brightness, 100), 0)
    brightness_value = int((brightness / 100) * 100)
    
    # Use WMI to set brightness
    c = wmi.WMI(namespace='wmi')
    methods = c.WmiMonitorBrightnessMethods()[0]
    methods.WmiSetBrightness(brightness_value, 0)

def get_brightness_win():
    return win32api.GetMonitorBrightness(win32api.EnumDisplayMonitors()[0][0])
cam = cv2.VideoCapture(0)
cam.set(3, wCam)
cam.set(4, hCam)
```

```python
with mp_hands.Hands(model_complexity=0, min_detection_confidence=0.5, min_tracking_confidence=0.5) as hands:

    while cam.isOpened():
        success, image = cam.read()

        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        results = hands.process(image)
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                mp_drawing.draw_landmarks(
                    image,
                    hand_landmarks,
                    mp_hands.HAND_CONNECTIONS,
                    mp_drawing_styles.get_default_hand_landmarks_style(),
                    mp_drawing_styles.get_default_hand_connections_style()
                )

            # multi_hand_landmarks method for Finding position of Hand landmarks
            lmList = []
            if results.multi_hand_landmarks:
                myHand = results.multi_hand_landmarks[0]
                for id, lm in enumerate(myHand.landmark):
                    h, w, c = image.shape
                    cx, cy = int(lm.x * w), int(lm.y * h)
                    lmList.append([id, cx, cy])

            # Volume and Brightness Control
            if len(lmList) != 0:
                x1, y1 = lmList[4][1], lmList[4][2]
                x2, y2 = lmList[8][1], lmList[8][2]

                # Marking Thumb and Index finger
                cv2.circle(image, (x1, y1), 15, (255, 255, 255))
                cv2.circle(image, (x2, y2), 15, (255, 255, 255))
                cv2.line(image, (x1, y1), (x2, y2), (0, 255, 0), 3)
                length = math.hypot(x2 - x1, y2 - y1)
                if length < 50:
                    cv2.line(image, (x1, y1), (x2, y2), (0, 0, 255), 3)

                # Volume Control (Left Half of the Screen)
                if x1 < wCam // 2 and x2 < wCam // 2:
                    vol = np.interp(length, [50, 220], [minVol, maxVol])
                    volume.SetMasterVolumeLevel(vol, None)
                    volBar = np.interp(length, [50, 220], [400, 150])
                    volPer = np.interp(length, [50, 220], [0, 100])

                # Brightness Control (Right Half of the Screen)
                if x1 >= wCam // 2 and x2 >= wCam // 2:
                    bright = np.interp(length, [50, 220], [minBright, maxBright])
                    bright = int(brightFactor * bright)  # Adjust brightness sensitivity
                    set_brightness_win(bright)
                    bright_rgb = np.full_like(image, (bright, bright, bright), dtype=np.uint8)
                    image = cv2.add(image, bright_rgb)
                    brightBar = np.interp(length, [50, 220], [150, 400])
                    brightPer = np.interp(length, [50, 220], [0, 100])

        # Draw a vertical line to separate the screen into two halves
        cv2.line(image, (wCam // 2, 0), (wCam // 2, hCam), (255, 255, 255), 2)

        # Draw volume and brightness bars on the screen
        cv2.rectangle(image, (50, 150), (85, 400), (20, 20, 20), 5)
        cv2.rectangle(image, (50, int(volBar)), (85, 400), (0, 0, 0), cv2.FILLED)
        cv2.putText(image, f'{int(volPer)} %', (40, 450), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 0), 3)

        cv2.rectangle(image, (950, 150), (985, 400), (20, 20, 20), 5)
        cv2.rectangle(image, (950, int(brightBar)), (985, 400), (0, 0, 0), cv2.FILLED)
        cv2.putText(image, f'{int(brightPer)} %', (955, 450), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 0), 3)

        cv2.imshow('handDetector', image)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cam.release()
cv2.destroyAllWindows()
```
